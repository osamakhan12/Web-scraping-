```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python

html = ('https://mufap.com.pk/aum_report.php?tab=01&fname=&amc=&cat=&strdate=&endate=&submitted=Show%20Report&mnt=01&yrs=2020&s=')
req = requests.get(html)
html_cont = req.content
soup = soup(html_cont,'html.parser')
data=[]

table = soup.find('table',class_="mydata")

try:
    for row in table.find_all('tr',class_=['tab-data border','tab-data border report-row-back']):
            cols = row.find_all('td')
            data.append((cols[0].text.strip(), cols[1].text.strip(), cols[2].text.strip(), cols[3].text.strip()))
except:
    "ok"
data
```

```python
df = pd.DataFrame(data)
df.columns = ['Fund_Name', 'Category','Inception_Date', 'Published']
```

```python
df['Published']=df['Published'].replace("Not Published","0")
```

```python
df['Inception_Date'] = pd.to_datetime(df['Inception_Date'])
```

```python
df['Published'] = df['Published'].str.replace(',', '')
```
